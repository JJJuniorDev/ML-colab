```python
import os
import tarfile
import urllib.request
import pandas as pd


train = pd.read_csv("train.csv")
test= pd.read_csv("test.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler

#Feature e target
X= train.drop(["Survived", "Name", "Ticket", "Cabin"], axis=1)
y = train["Survived"].astype(int)

#ENCODING COLONNE CATEGORICHE
X["Sex"] = X["Sex"].map({"male": 0, "female": 1})
X["Embarked"] = X["Embarked"].map({"S": 0, "C": 1, "Q": 2})

# --- RIEMPIMENTO NaN (minimo indispensabile) ---
X["Age"] = X["Age"].fillna(X["Age"].median())
X["Embarked"] = X["Embarked"].fillna(X["Embarked"].mode()[0])
X["Fare"] = X["Fare"].fillna(X["Fare"].median())

# --- STANDARDIZZAZIONE ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sgd_clf= SGDClassifier(random_state=42)

#sgd_clf.fit(X,y)
y_train_pred= cross_val_predict(sgd_clf, X_scaled, y, cv=3)
cm=confusion_matrix(y, y_train_pred)
print(cm)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision=precision_score(y, y_train_pred)
print("PRECISIONE ", precision)
recall=recall_score(y, y_train_pred)
print("RECALL ", recall)
f1=f1_score(y, y_train_pred)
print("F1 ",f1)
```

STESSO CON LOGISTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score
#Feature e target
X= train.drop(["Survived", "Name", "Ticket", "Cabin"], axis=1)
y = train["Survived"].astype(int)

#ENCODING COLONNE CATEGORICHE
X["Sex"] = X["Sex"].map({"male": 0, "female": 1})
X["Embarked"] = X["Embarked"].map({"S": 0, "C": 1, "Q": 2})

# --- RIEMPIMENTO NaN (minimo indispensabile) ---
X["Age"] = X["Age"].fillna(X["Age"].median())
X["Embarked"] = X["Embarked"].fillna(X["Embarked"].mode()[0])
X["Fare"] = X["Fare"].fillna(X["Fare"].median())

# --- STANDARDIZZAZIONE ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

log_reg= LogisticRegression(max_iter=1000, random_state=42)

#sgd_clf.fit(X,y)
y_train_pred= cross_val_predict(log_reg, X_scaled, y, cv=3)
cm=confusion_matrix(y, y_train_pred)
print(cm)

precision=precision_score(y, y_train_pred)
print("PRECISIONE ", precision)
recall=recall_score(y, y_train_pred)
print("RECALL ", recall)
f1=f1_score(y, y_train_pred)
print("F1 ",f1)
```

RANDOM FOREST


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score
#Feature e target
X= train.drop(["Survived", "Name", "Ticket", "Cabin"], axis=1)
y = train["Survived"].astype(int)

#ENCODING COLONNE CATEGORICHE
X["Sex"] = X["Sex"].map({"male": 0, "female": 1})
X["Embarked"] = X["Embarked"].map({"S": 0, "C": 1, "Q": 2})

# --- RIEMPIMENTO NaN (minimo indispensabile) ---
X["Age"] = X["Age"].fillna(X["Age"].median())
X["Embarked"] = X["Embarked"].fillna(X["Embarked"].mode()[0])
X["Fare"] = X["Fare"].fillna(X["Fare"].median())

#NON SERVE SCALING!!
# --- RANDOM FOREST BASELINE ---
rf_clf = RandomForestClassifier(
    random_state=42
)
y_train_pred= cross_val_predict(rf_clf, X, y, cv=3)
print("Baseline F1:", f1_score(y, y_train_pred))


# --- GRID SEARCH ---
param_grid = {
    "n_estimators": [50, 100, 200],
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None]
}

grid_search = GridSearchCV(
    rf_clf,
    param_grid,
    cv=3,
    scoring="f1",
    n_jobs=-1
)
grid_search.fit(X, y)
print("Best Grid F1:", grid_search.best_score_)
print("Best Grid Params:", grid_search.best_params_)
```
